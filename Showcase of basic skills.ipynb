# Using plots in pandas

```python
import pandas as pd
```

```python
df_population_row=pd.read_csv(r"C:\Users\User\Downloads\population_total.csv")
```

```python
df_population_row
```

```python
#dropping null values
df_population_row.dropna(inplace=True)
```

```python
df_population_row
```

```python
# making pivot table
df_pivot=df_population_row.pivot(index='year', columns='country', values='population')
```

```python
df_pivot
```

```python
# selecting some countries
df_pivot=df_pivot[['United States', 'China', 'India', 'Nigeria', 'Russia']]
```

```python
df_pivot
```

```python
import matplotlib.pyplot as plt
```

### Making line plot

```python
df_pivot.plot(kind='line',xlabel='Year', ylabel='Population', title="POPULATION FROM 1955 to 2020",
             figsize=(8,4))
# save fig
plt.savefig('my_test.png')
# show plot
plt.show()
```

# Barplot
